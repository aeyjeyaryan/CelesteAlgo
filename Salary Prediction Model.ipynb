```python
import pandas as pd
import numpy as np
import seaborn as sns
ages = np.random.randint(20, 61, 100)

salaries = 50000 + 1000 * ages + np.random.normal(0, 10000, 100)  #this block adds randomness to the salary of the peopl with different ages and
#it also randomized the salary based on the age!

df = pd.DataFrame({'Age': ages, 'Salary': salaries})

print(df.head())
```

```python
sns.lmplot(x='Age',y='Salary',data=df,hue='Salary')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
x = np.asanyarray(df[['Salary']])
y = np.asanyarray(df[['Age']])
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.5)
```

```python
import math
math.sqrt(len(ytest))
```

```python
knn = KNeighborsRegressor()
knn.fit(xtrain,ytrain)
```

```python
knn = KNeighborsRegressor (n_neighbors = 7)
knn.fit(xtrain, ytrain)
pred = knn.predict(xtest)
```

```python
inp = float(input("Enter your age: "))
features = np.array([[inp]])
predicted_salary = knn.predict(features)
print("Based on your age, your predicted salary is: ", predicted_salary)
```

```python
knn.score(x,y)
```
