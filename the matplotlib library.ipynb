```python
import matplotlib.pyplot as plt
```

```python
from matplotlib import style
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
x
```

```python
# functional
style.use('ggplot')
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Progress')
plt.title("Progress")
#plt.show() for other interpreters
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('Time')
axes.set_ylabel("Work")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger")
axes1.set_xlabel("X")
axes1.set_ylabel("Y")

axes2.plot(y,x)
axes2.set_title("smaller")
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
# for z in axes:
#   z.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)


axes[0].set_title("first plot")
axes[1].set_title("second plot")
axes[0].set_xlabel("Xlabel")
axes[1].set_ylabel("Ylabel")
```

```python
#figure size and dpi(dots per inch)
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig("mypic.png",dpi=2000)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Square')
ax.plot(x,x**3,label='Y Square')

# ax.set_title("title")
# ax.set_ylabel("Y")
# ax.set_xlabel("X")

ax.legend(loc=10)

#parameters to arrange the legend on a graph
# bes - 1
# upper right - 1
# upper left - 2
# lower left - 3
# lower right - 4
# right - 5
# center left - 6
# center right - 7
# lower center - 8
# upper center - 9
# center - 10
```

```python
# plot apperance
#issue : the colors are not changing
x1 = np.linspace(0,5,11)
y1 = x1 ** 2
gif = plt.figure()

ad = gif.add_axes([0,0,1,1])  #lw - linewidth ls - linestyle
ad.plot(x1,y1,color="orange",lw=2,ls="dotted")#,marker="+",markersize=45,
       # markerfacecolor="blue",markeredgewidth=3,markeredgecolor='green')   #alpha=x.x for the line transparency

ad.set_xlim([0,1])
ad.set_ylim([0,2])
```
