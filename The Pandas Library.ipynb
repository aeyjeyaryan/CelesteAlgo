```python
#missing data
mydict = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
pdf = pd.DataFrame(mydict)
pdf
```

```python
pdf.dropna()   #this deletes the rows that has atleast one null value
#and axis=1 for column for sure
```

```python
pdf.dropna(thresh=2)  #if there is 2 nan values in any row then it gets deleted if set thresh = 3 then row with three null value gets dropeed
```

```python
#to fill null values we use fillna
pdf.fillna(value="aryan")
```

```python
#to set the mean value in place of any null value
pdf['B'].fillna(value=pdf['A'].mean())
```

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
pd.merge(left,right,how='outer',on='key')   #same output on both inner and outer
```

```python
left = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                     'key2':['K0','K1','K0','K1'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'key1':['K0','K1','K1','K2'],
                      'key2':['K0','K0','K0','K0'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
pd.merge(left,right,on=['key1','key2'])
```

```python
pd.merge(left,right,how='outer',on=['key1','key2'])
```

```python
pd.merge(left,right,how='right',on=['key1','key2'])
```
